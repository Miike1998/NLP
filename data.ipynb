```python
from datasets import load_dataset
from transformers import pipeline
import torch
import numpy as np

# 1. Laad Engelse dataset
ds = load_dataset("google-research-datasets/go_emotions", "simplified")

# 2. Vertaalmodel
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-nl")

# 3. Functie om tekst te vertalen
def translate_batch(examples):
    translations = translator(examples["text"])
    examples["text_nl"] = [t["translation_text"] for t in translations]
    return examples

# 4. Toepassen op de volledige dataset
ds_nl = ds.map(translate_batch, batched=True)

# 5. Eerste 5 vertaalde waardes
print(ds_nl["train"][0]["text"])
print(ds_nl["train"][0]["text_nl"])
```

```python
from transformers import pipeline

translator = pipeline(
    "translation",
    model="facebook/mbart-large-50-many-to-many-mmt"
)

result = translator(
    "Hello, how are you?",
    src_lang="en_XX",
    tgt_lang="nl_XX"
)

print(result[0]["translation_text"])
```

```python
def translate_batch(batch):
    outputs = translator(
        batch["text"],
        src_lang="en_XX",
        tgt_lang="nl_XX"
    )
    batch["text_nl"] = [o["translation_text"] for o in outputs]
    return batch
```

```python
from datasets import load_dataset
ds = load_dataset("google-research-datasets/go_emotions", "simplified")

ds_nl = ds.map(translate_batch, batched=True)
```
